# Scatter

```python
plt.scatter(a['Price'],a['Age_08_04'],c='red')
plt.title('Scatter plot of Price vs Age of the cars')
plt.xlabel('Age_08_04(months)')
plt.ylabel('Price(Euros)')
plt.show()
```

# Histogram

```python
plt.hist(a['KM'])
```

# Histogram with space

```python
plt.hist(a['KM'],color='green',edgecolor = 'orange',bins=5)
plt.title('Histogram of kilometer')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
        
```

# Bar Plot

```python
counts = [979,120,12]
fuelType = ('Petrol', 'Diesel','CNG')
index = np.arange(len(fuelType))
plt.bar(index,counts, color=['red','blue','orange'])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.show()
```

```python
counts = [979,120,12]
fuelType = ('Petrol', 'Diesel','CNG')
index = np.arange(len(fuelType))
plt.bar(index,counts, color=['green','cyan','orange'])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index, fuelType,rotation =90)
plt.show()
```
